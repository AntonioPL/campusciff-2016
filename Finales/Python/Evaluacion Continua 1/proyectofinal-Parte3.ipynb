```python
# Trabajo final de Python
import pandas as pd
import numpy as np
import datetime 
import statsmodels.api as sm
```

```python
Forbes2000 = sm.datasets.get_rdataset('Forbes2000', 'HSAUR')
Forbes2000 = Forbes2000.data
```

```python
Forbes2000.head(10)
```

```python
# Selecionar del dataframe los que incluyan "elect", no se incluye Mayuscula
# (sin tener en cuenta may√∫sculas), segun entiendo, hay que excluir mayusculas, pero si es asi, no se puede continuar
# con los siguientes pasos, asique se incluyen Mayusculas

Forbes2000red = Forbes2000[Forbes2000['name'].str.contains("elect")==True]
```

```python
Forbes2000red.head(10)
```

```python
# Selecionar del dataframe los que incluyan "elect", se incluye E Mayuscula
Forbes2000ele = Forbes2000[Forbes2000['name'].str.contains('elect', case=False)==True]
```

```python
Forbes2000ele.head(10)
```

```python
# Utilizando como √≠ndice el pa√≠s y la categor√≠a y orden√°ndolo (sort_index), encontrar el
# valor de mercado de las empresas japonesas y estadounidenses.
# la funcion sort_index esta depreciada, por lo que se sugiere utilizar .sort_values

# Efectuando Sort
ForUSAJP = Forbes2000ele[Forbes2000ele['country'].isin(['United States', 'Japan'])].sort_values(by=['country', 'category'])
```

```python
# Resultado Sort
ForUSAJP
```

```python
# Valor de las empresas USA y de Japan

MkvakUSAJP = ForUSAJP.sort_values(["country","category"], ascending=False).groupby("country", as_index=False).sum()
MkvakUSAJP.loc[:, ['country','marketvalue']]
```

```python
# Sobre el mismo subconjunto de empresas el√©ctricas, obtener el pa√≠s donde el sector de
# utilities tiene mayor valor de mercado medio en sus empresas

# Efectuando Sort
CountryUtil = Forbes2000ele[Forbes2000ele['category'].isin(['Utilities'])].sort_values(by=['country'])
```

```python
CountryUtil
```

```python
MaxCountryUtil = CountryUtil.sort_values(['country'], ascending=False).groupby("country", as_index=False).mean()
```

```python
#CountryUtil.loc[:, ['country','marketvalue']]
MaxCountryUtil
```

```python
MaxCountryUtil.drop(['rank','sales','profits', 'assets'], axis=1, inplace=True)
MaxCountryUtil.ix[MaxCountryUtil['marketvalue'].idxmax()]
```

```python
# Volviendo al dataset original, obtener las empresas espa√±olas
Forbes2000ESP = Forbes2000[Forbes2000.country == 'Spain']
```

```python
Forbes2000ESP
```

```python
# Indexar el dataframe anterior por beneficios (profits) y obtener los nombres y rangos de
# las que tengan un beneficio nulo o negativo. Nota: intenta indexar de manera mixta con
# n√∫meros (filas) y etiquetas columnas), hay una operaci√≥n espec√≠fica para esto en pandas

# Index por Profits

Forbes2000ESPidx = Forbes2000ESP.set_index('profits')
```

```python
Forbes2000ESPidx.head(3)
```
