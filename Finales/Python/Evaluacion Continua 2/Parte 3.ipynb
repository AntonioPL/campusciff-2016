# Opcion archivos TXT
Se leearan todos los archivos del 2015 y se concatenaran
(Se cambian archivos "precip.txt" por "dialy.txt", ya que los primero parecen tener alguna informacion incorrecta.
Se hace un FOR iniciando con el archivo de fecha mas antigua  y se recorre mes a mes, obligando a que se cargen secuancialmente.

```python
dateInicio = 201400
Meses = 33
dateini = dateInicio
fileop = ''
yearCO = 0
CreaDF = 1
for cont in range(Meses):
    dateini = dateini + 1
    yearCO = yearCO + 1
    fileop = './NOOA/'+ str(dateini) + 'daily.txt'
    print "Operando archivo -- > " + fileop
    %time Tprecip = pd.read_csv(fileop, sep=',', header=0, converters={'YearMonthDay':str})
    if CreaDF == 1:
        precip = Tprecip
        CreaDF = 0
    else:
        precip = pd.concat([precip, Tprecip], ignore_index=True)
    if yearCO == 12:
        yearCO = 0
        dateini = dateini + 88
```

Creamos campo Año (Servicer para facilitar un nuevo calculo)

```python
yearca = precip.YearMonthDay.str[:4]
precip['Year'] = yearca
precip.head(15)
```

Hay campos que vienen con valores que NO SON NUMERICOS, por lo que le damos valor CERO
de lo contrario no podremos hacer el calculo de validar los dias de mayor lluvia

```python
precip.PrecipTotal.replace(['M','  T','err'], ['0.00','0.00','0.00'], inplace=True)
precip['PrecipTotal'] = precip['PrecipTotal'].astype('float') 
```

# Determinaremos el dia que hay mas precipitaciones y lo mostramos

```python
%time Masprecip = precip.groupby('YearMonthDay').agg({'PrecipTotal': lambda x: x.max()})
Masprecip.head()
```

Reseteamos el indice y creamos uno nuevo, calculamos el registro que tiene el DIA con mayor lluvia y lo mostramos

```python
Masprecip = Masprecip.reset_index(drop=False)
ind = Masprecip['PrecipTotal'].idxmax()
print Masprecip.ix[ind]
```
